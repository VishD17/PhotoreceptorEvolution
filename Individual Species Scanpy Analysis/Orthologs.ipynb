```python
#Loading ortholog data
import pandas as pd
df = pd.read_csv("/Users/vishruthdinesh/Downloads/martMergeRefHuman.csv")
```

```python
#Subsetting orthologous gene dataset to only include species used in analysis
oneToOneMartMouMarChiLiz = df[
    (df['Mouse homology type'] == 'ortholog_one2one') &
    (df['White-tufted-ear marmoset homology type'] == 'ortholog_one2one') &
    (df['Green anole homology type'] == 'ortholog_one2one')&
    (df["Chicken homology type"]=='ortholog_one2one')&
    (df["Cow homology type"]=='ortholog_one2one')&
     (df["Squirrel homology type"]=="ortholog_one2one")]
```

```python
oneToOneMartMouMarChiLiz
```

```python
max_nan_threshold = 4  # Define the maximum number of allowed NaNs per row

# Check how many NaN values in each row and filter rows with less than or equal to max_nan_threshold NaNs
df_cleaned = oneToOneMartMouMarChiLiz[oneToOneMartMouMarChiLiz.isna().sum(axis=1) <= max_nan_threshold]
```

```python
nan_count = oneToOneMartMouMarChiLiz.isna().sum(axis=1)


rows_with_nan = len(nan_count[nan_count > 0])

print(rows_with_nan)
```
